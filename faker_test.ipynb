```python
from collections import OrderedDict
from faker import Faker

locale_weights = OrderedDict([
    ("nl_NL", 0.82),  # Dutch / 2nd-gen Surinamese, Indonesian, etc.
    ("tr_TR", 0.04),  # Turkish origin
    ("ar_EG", 0.04),  # Moroccan + other Arabic/MENA (approximation)
    ("pl_PL", 0.025), # Polish
    ("de_DE", 0.02),  # German
    ("en_GB", 0.015), # British / Anglophone expats
    ("fr_FR", 0.01),  # French / some North-African mix
    ("es_ES", 0.01),  # Southern EU migrants
    ("it_IT", 0.01),  # Southern EU migrants
    ("ro_RO", 0.01),  # Eastern EU migrants
])

fake = Faker(locale=locale_weights, use_weighting=True)
```

```python
print(fake.name())
print("\n")
print(fake.address())
```

```python
from faker import Faker
from faker.providers import DynamicProvider

medical_professions_provider = DynamicProvider(
     provider_name="medical_profession",
     elements=["dr.", "doctor", "nurse", "surgeon", "clerk"],
)

fake = Faker()

# then add new provider to faker instance
fake.add_provider(medical_professions_provider)

# now you can use:
fake.medical_profession()
```
